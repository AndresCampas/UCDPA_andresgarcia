```python
def size_population(x):
    if x < 5:
        return 'Very Small'
    elif x < 10:
        return 'Small'
    elif x < 30: 
        return'Medium'
    elif x < 50:
        return 'Medium-Big'
    elif x <100:
        return 'Big'
    else:
        return 'Very Big'

Europe_eco_data['Size Of Population'] = Europe_eco_data['Population (Millions)'].apply(size_population)

# We create a custom function that use conditional statments. 
```
